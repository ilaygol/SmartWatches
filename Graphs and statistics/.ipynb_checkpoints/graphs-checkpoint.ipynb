```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
pd.set_option("display.precision", 2)
plt.rcParams.update({'font.size': 20, 'figure.figsize': (8, 4)})

%matplotlib inline
import matplotlib_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('svg')

import seaborn as sns
sns.set()
```

```python
df = pd.read_csv("measurements.csv")
df['numeric_id'] = df['id'].apply(lambda id: int(id))
df.info()

morning_df = df[df["numeric_id"]<200]
night_df = df[df["numeric_id"]>=200]
```

```python
average_steps_morning_df = morning_df.groupby('id')['steps'].mean().to_frame()
average_steps_night_df = night_df.groupby('id')['steps'].mean().to_frame()
average_morning_steps = average_steps_morning_df['steps'].mean()
average_night_steps = average_steps_night_df['steps'].mean()
display(average_steps_morning_df)
display(average_steps_night_df)
"""
data = {'type':['morning-type', 'night-type'], 'average steps':[average_morning_steps, average_night_steps]}
average_steps_df = pd.DataFrame(data)
ax = average_steps_df.plot.bar(x='type' , y='average steps' ,title='average steps per type')
for container in ax.containers:
    ax.bar_label(container)
average_steps_df
"""
```

```python
average_heart_rate_morning_df = morning_df.groupby('id')['average heartrate'].mean().to_frame()
average_heart_rate_night_df = night_df.groupby('id')['average heartrate'].mean().to_frame()
average_morning_heartrate = average_heart_rate_morning_df['average heartrate'].mean()
average_night_heartrate = average_heart_rate_night_df['average heartrate'].mean()

data = {'type':['morning-type', 'night-type'], 'average heartRate':[average_morning_heartrate, average_night_heartrate]}
average_heartrate_df = pd.DataFrame(data)
average_heartrate_df.plot.bar(x='type' , y='average heartRate' ,title='average heartRate per type')
```

```python
average_stress_morning_df = morning_df.groupby('id')['average stress'].mean().to_frame()
average_stress_night_df = night_df.groupby('id')['average stress'].mean().to_frame()
average_morning_stress = average_stress_morning_df['average stress'].mean()
average_night_stress = average_stress_night_df['average stress'].mean()

data = {'type':['morning-type', 'night-type'], 'average stress':[average_morning_stress, average_night_stress]}
average_stress_df = pd.DataFrame(data)

average_stress_df.plot.bar(x='type' , y='average stress' ,title='average stress per type')
```

```python
merged_df = pd.merge(average_stress_df, average_heartrate_df)
merged_df.plot.bar(x='type', title='morning vs night type')
merged_df
```

```python
data2 = {'average':['Average stress', 'Average-heartRate'], 'morning-type':[average_morning_stress, average_morning_heartrate], 'night-type':[average_night_stress, average_night_heartrate]}
merged_df_2 = pd.DataFrame(data2)
ax = merged_df_2.plot(kind='bar', x='average', title='morning vs night types')
for container in ax.containers:
    ax.bar_label(container)
merged_df_2
```

```python
merged1 = pd.merge(average_steps_morning_df.reset_index(), average_heart_rate_morning_df.reset_index())
merged2 = pd.merge(merged1, average_stress_morning_df.reset_index(), on='id').set_index('id')
merged2.to_csv('morning-type-averages.csv', encoding='utf-8', float_format = '%.2f')
merged2
```
